# Parameter sweep

#### Note: Code to perform the parameter sweep is in `klic_parameter_sweep.py`

```python
import os
import time
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns


%matplotlib inline
%load_ext autoreload
%autoreload 2
```

# Load in data from parameter sweep and plot figures

```python
df = pd.read_csv('results/klic_parameter_sweep.csv')
```

```python
index = 'Clusters'
columns = 'Scale'

plt.figure(figsize=(3.25,3))
for i, g in enumerate(df.Gamma.unique()):
    plt.subplot(1,3,i+1)
    tmpdf = df.loc[df.Gamma==g]
    pivot = pd.pivot_table(tmpdf, values='Loss', index=index, columns=columns, aggfunc='mean', fill_value=0)
    if i==2:
        ax = sns.heatmap(data=pivot,cmap='seismic',center=0,vmin=0, vmax=3, square=True,cbar_kws={'fraction':0.046})
        colorbar = ax.collections[0].colorbar
        colorbar.ax.tick_params(labelsize=6) # Change font size for both major and minor ticks
    else:
        ax = sns.heatmap(data=pivot,cmap='seismic',center=0,vmin=0, vmax=3, square=True,cbar=False)

    if i != 0:
        plt.ylabel(None)


    ax.invert_yaxis()
    ax.xaxis.label.set_size(8)
    ax.yaxis.label.set_size(8)
    plt.yticks(np.arange(0,len(pivot.index),2),pivot.index[::2])
    plt.xlabel('Scaling',fontsize=8)
    ax.tick_params(labelsize=6)
    plt.title(r"$\gamma=$"+f"{g}",fontsize=8)
    plt.tight_layout()
    plt.savefig(f"results/{index}_v_{columns}.pdf",transparent=True)
```
